## Observation:

* In the above graph we can see a very high imbalance in the data- i.e. The bankrupt companies are very less(only 3.2%) as compared to non bankrupt.


* We need to handle this imbalance so as to get a proper prediction and an efficient model, otherwise it is possible that our model will predict every company to be not likely to be bankrupt.


* To handle the imbalance we will use the SMOTE technique.

#### We will use the technique after splitting the  data into train test, and apply the SMOTE only on train data.

* If we apply the SMOTE directly on whole data without the split, you are essentially creating the data that will reflect in your testing set as well(the sample will be present after the splitting) which will provide misleading results.


* SMOTE is only synthetic oversampling and the samples should not be used as original data points.

## Let us get a clearer view of the correlation:

```python
cor_matrix = df.corr().abs()
cor_matrix.style.background_gradient(sns.light_palette('skyblue', as_cmap = True))
```

## Further Plan of Actions:

* I will try to drop the correlated features which have more than 0.9 corr coefficient.


* Will use SMOTE to balance the dataset


* Use various classification algorithm to test and check the results( this code will contain the final output, I will try to tune hyperparameters manually for different algorithms for the sake of learning myself).


* I will print all the confusion matrices at the end of the code to look and finally write the final observation of the model.

* I will not use accuracy as the judging metric because of the high imbalance in the dataset rather I will use 
1. F1 score
2. Precision
3. Recall 

and most importantly,

4. AUC/ ROC 

```python
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool)) #upper triangle of the correlation matrix

dropped_cols = set()
for feature in upper_tri.columns:
    if any(upper_tri[feature] > 0.9): #more than 0.9 corr. coeficient -> dropped
        dropped_cols.add(feature)

print("There are %d dropped columns" %len(dropped_cols))

df = df.drop(dropped_cols,axis=1)
df.head()
```

### Applying the train test split to the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,stratify= y, random_state=35)
```

### Importing SMOTE and applying the technique

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE()
```

```python
x_train_sm, y_train_sm = sm.fit_resample(x_train, y_train.ravel())
```
